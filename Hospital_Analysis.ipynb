## 📊 Step 4 — Data Visualizations

```python
def add_stamp(ax):
    """Add student info watermark to every chart."""
    ax.text(0.99, 0.01, 'Syed Shaban | SAP: 80414 | DAA | Sir Karar Haider',
            transform=ax.transAxes, fontsize=7.5, color='#555555',
            ha='right', va='bottom', alpha=0.75,
            bbox=dict(boxstyle='round,pad=0.2', facecolor='white', alpha=0.5))

print('✅ add_stamp() helper defined — every chart will carry your name.')
```

```python
# ── Chart 1 : Bed Capacity Comparison ────────────────────────────
fig, ax = plt.subplots(figsize=(14, 8))
colors = ['#e63946' if b >= 1500 else '#1a1a2e' for b in df['Beds']]
bars = ax.barh(df['Hospital_Name'], df['Beds'], color=colors, edgecolor='white', linewidth=0.5)
ax.set_xlabel('Number of Beds', fontsize=12)
ax.set_title('🏥  Hospitals — Bed Capacity Comparison\nSyed Shaban | SAP: 80414 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold', pad=12)
ax.axvline(1500, color='#e63946', linestyle='--', alpha=0.5, label='1500 Bed Mark')
for bar, val in zip(bars, df['Beds']):
    ax.text(val + 15, bar.get_y() + bar.get_height() / 2,
            f'{val}', va='center', fontsize=8)
ax.legend(); add_stamp(ax); plt.tight_layout()
plt.savefig('chart1_bed_capacity.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 1 saved.')
```

```python
# ── Chart 2 : Budget vs Annual Patients ──────────────────────────
fig, ax = plt.subplots(figsize=(12, 6))
sc = ax.scatter(df['Annual_Patients'] / 1000, df['Annual_Budget_USD'] / 1e9,
                c=df['Founded'], cmap='plasma', s=110, edgecolors='white', linewidth=0.8)
for _, row in df.iterrows():
    ax.annotate(row['Hospital_Name'], (row['Annual_Patients'] / 1000, row['Annual_Budget_USD'] / 1e9),
                textcoords='offset points', xytext=(6, 4), fontsize=7, alpha=0.85)
plt.colorbar(sc, ax=ax, label='Founded Year')
ax.set_xlabel('Annual Patients (Thousands)', fontsize=12)
ax.set_ylabel('Annual Budget (Billion USD)', fontsize=12)
ax.set_title('💰  Budget vs Annual Patients\nSyed Shaban | SAP: 80414 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold')
add_stamp(ax); plt.tight_layout()
plt.savefig('chart2_budget_vs_patients.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 2 saved.')
```

```python
# ── Chart 3 : Top 10 Hospitals by Annual Patients ────────────────
top10 = df.nlargest(10, 'Annual_Patients')
fig, ax = plt.subplots(figsize=(12, 6))
bars = ax.bar(top10['Hospital_Name'], top10['Annual_Patients'] / 1000,
              color=sns.color_palette('rocket', len(top10)), edgecolor='white', linewidth=0.6)
ax.set_ylabel('Annual Patients (Thousands)', fontsize=12)
ax.set_title('👩‍⚕️  Top 10 Hospitals by Annual Patients\nSyed Shaban | SAP: 80414 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold')
plt.xticks(rotation=38, ha='right', fontsize=9)
for bar, val in zip(bars, top10['Annual_Patients']):
    ax.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 20,
            f'{val/1000:.0f}k', ha='center', fontsize=8)
add_stamp(ax); plt.tight_layout()
plt.savefig('chart3_top_patients.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 3 saved.')
```

```python
# ── Chart 4 : Staff Strength — Doctors vs Nurses ─────────────────
ms = df.sort_values('Founded')
fig, ax = plt.subplots(figsize=(13, 6))
ax.plot(ms['Hospital_Name'], ms['Doctors'], marker='o', color='#e63946',
        linewidth=2.2, markersize=7, markerfacecolor='white', markeredgewidth=2, label='Doctors')
ax.plot(ms['Hospital_Name'], ms['Nurses'], marker='s', color='#1a1a2e',
        linewidth=2.2, markersize=7, markerfacecolor='white', markeredgewidth=2, label='Nurses')
ax.set_xlabel('Hospital (ordered by founding year)', fontsize=12)
ax.set_ylabel('Staff Count', fontsize=12)
ax.set_title('📈  Staff Strength — Doctors vs Nurses\nSyed Shaban | SAP: 80414 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=12, fontweight='bold')
plt.xticks(rotation=45, ha='right', fontsize=8)
ax.legend(); add_stamp(ax); plt.tight_layout()
plt.savefig('chart4_staff_strength.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 4 saved.')
```

```python
# ── Chart 5 : Correlation Heatmap ────────────────────────────────
num = df[['Beds','ICU_Beds','Doctors','Nurses','Annual_Patients','Specialties','Rating_Out_Of_5','Annual_Budget_USD']]
corr = num.corr()
fig, ax = plt.subplots(figsize=(10, 7))
mask = np.triu(np.ones_like(corr, dtype=bool))
sns.heatmap(corr, mask=mask, annot=True, fmt='.2f', cmap='coolwarm',
            linewidths=0.5, ax=ax, vmin=-1, vmax=1)
ax.set_title('🔗  Correlation Matrix — Capacity, Staff & Budget\nSyed Shaban | SAP: 80414 | Design and Analysis of Algorithms | Sir Karar Haider',
             fontsize=11, fontweight='bold')
add_stamp(ax); plt.tight_layout()
plt.savefig('chart5_correlation.png', dpi=150, bbox_inches='tight')
plt.show()
print('✅ Chart 5 saved.')
```
